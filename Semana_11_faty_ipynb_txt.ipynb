````python


 def info_general(df_estudiantes: pd.DataFrame, df_calificaciones: pd.DataFrame) -> Dict:

    return {
        "total_estudiantes": len(df_estudiantes),
        "total_registros_calif": len(df_calificaciones),
        "semestres": sorted(df_estudiantes["semestre"].unique().tolist()),
        "materias_con_registros": df_calificaciones["materia_id"].nunique()
    }
````

````python
def validar_datos(df_calificaciones: pd.DataFrame) -> Dict:

    registros_con_nulos = df_calificaciones.isna().any(axis=1).sum()

    fuera_rango = (
        (df_calificaciones["parcial_1"] < 0) |
        (df_calificaciones["parcial_1"] > 10) |
        (df_calificaciones["parcial_2"] < 0) |
        (df_calificaciones["parcial_2"] > 10) |
        (df_calificaciones["final"] < 0) |
        (df_calificaciones["final"] > 10)
    ).sum()

    return {
        "registros_con_nulos": int(registros_con_nulos),
        "calificaciones_fuera_rango": int(fuera_rango),
        "datos_validos": registros_con_nulos == 0 and fuera_rango == 0
    }
````

````python
# PARTE 2: Consultas y Filtros

def buscar_estudiante(df_estudiantes: pd.DataFrame,
                      criterio: str,
                      valor: str) -> pd.DataFrame:

    if criterio == "boleta":
        return df_estudiantes[df_estudiantes["boleta"] == valor]

    elif criterio == "nombre":
        return df_estudiantes[
            df_estudiantes["nombre"].str.contains(valor, case=False)
        ]

    elif criterio == "semestre":
        return df_estudiantes[
            df_estudiantes["semestre"] == int(valor)
        ]

    return pd.DataFrame()
````

````python
def obtener_kardex(boleta: str,
                   df_estudiantes: pd.DataFrame,
                   df_calificaciones: pd.DataFrame,
                   df_materias: pd.DataFrame) -> Dict:

    estudiante = df_estudiantes[df_estudiantes["boleta"] == boleta]

    if estudiante.empty:
        return {}

    calif = df_calificaciones[
        df_calificaciones["boleta"] == boleta
    ].copy()

    materias = calif.merge(
        df_materias,
        on="materia_id",
        how="left"
    )

    materias["promedio"] = materias[
        ["parcial_1", "parcial_2", "final"]
    ].mean(axis=1)

    return {
        "estudiante": estudiante.iloc[0].to_dict(),
        "materias": materias,
        "promedio_general": materias["promedio"].mean(),
        "creditos_cursados": materias["creditos"].sum(),
        "materias_aprobadas": int((materias["promedio"] >= 6).sum()),
        "materias_reprobadas": int((materias["promedio"] < 6).sum())
    }
````

````python
def filtrar_por_rendimiento(df_calificaciones: pd.DataFrame,
                            df_estudiantes: pd.DataFrame,
                            min_promedio: float = None,
                            max_promedio: float = None) -> pd.DataFrame:

    df = df_calificaciones.copy()

    df["promedio"] = df[
        ["parcial_1", "parcial_2", "final"]
    ].mean(axis=1)

    promedios = (
        df.groupby("boleta")["promedio"]
        .mean()
        .reset_index()
    )

    if min_promedio is not None:
        promedios = promedios[
            promedios["promedio"] >= min_promedio
        ]

    if max_promedio is not None:
        promedios = promedios[
            promedios["promedio"] <= max_promedio
        ]

    return promedios.merge(df_estudiantes, on="boleta")
````

````python
def calcular_promedio_materia(df_calificaciones: pd.DataFrame,
                              materia_id: str) -> Dict:

    materia = df_calificaciones[
        df_calificaciones["materia_id"] == materia_id
    ]

    promedios = materia[
        ["parcial_1", "parcial_2", "final"]
    ].mean(axis=1)

    return {
        "materia": materia_id,
        "inscritos": len(materia),
        "promedio_parcial1": materia["parcial_1"].mean(),
        "promedio_parcial2": materia["parcial_2"].mean(),
        "promedio_final": materia["final"].mean(),
        "promedio_general": promedios.mean(),
        "tasa_aprobacion": (promedios >= 6).mean() * 100,
        "calificacion_maxima": promedios.max(),
        "calificacion_minima": promedios.min()
    }
````

````python
def ranking_estudiantes(df_calificaciones: pd.DataFrame,
                        df_estudiantes: pd.DataFrame,
                        top_n: int = 10) -> pd.DataFrame:

    df = df_calificaciones.copy()

    df["promedio"] = df[
        ["parcial_1", "parcial_2", "final"]
    ].mean(axis=1)

    ranking = (
        df.groupby("boleta")["promedio"]
        .mean()
        .reset_index()
        .sort_values("promedio", ascending=False)
        .head(top_n)
    )

    return ranking.merge(df_estudiantes, on="boleta")
````

````python
def estadisticas_por_semestre(df_estudiantes: pd.DataFrame,
                              df_calificaciones: pd.DataFrame) -> pd.DataFrame:

    df = df_calificaciones.copy()

    df["promedio"] = df[
        ["parcial_1", "parcial_2", "final"]
    ].mean(axis=1)

    promedios = (
        df.groupby("boleta")["promedio"]
        .mean()
        .reset_index()
    )

    promedios = promedios.merge(
        df_estudiantes[["boleta", "semestre"]],
        on="boleta"
    )

    return promedios.groupby("semestre").agg(
        numero_estudiantes=("boleta", "count"),
        promedio_general=("promedio", "mean"),
        mejor_promedio=("promedio", "max"),
        peor_promedio=("promedio", "min")
    )
````

````python
def identificar_estudiantes_riesgo(df_calificaciones: pd.DataFrame,
                                   df_estudiantes: pd.DataFrame,
                                   umbral_promedio: float = 7.0,
                                   max_reprobadas: int = 2) -> pd.DataFrame:

    df = df_calificaciones.copy()

    df["promedio"] = df[
        ["parcial_1", "parcial_2", "final"]
    ].mean(axis=1)

    promedio_general = (
        df.groupby("boleta")["promedio"]
        .mean()
        .reset_index()
    )

    reprobadas = (
        df.groupby("boleta")
        .apply(lambda x: (x["promedio"] < 6).sum())
        .reset_index(name="materias_reprobadas")
    )

    riesgo = promedio_general.merge(reprobadas, on="boleta")

    riesgo = riesgo[
        (riesgo["promedio"] < umbral_promedio) |
        (riesgo["materias_reprobadas"] > max_reprobadas)
    ]

    return riesgo.merge(df_estudiantes, on="boleta")
````

````python
def generar_reporte_academico(df_estudiantes: pd.DataFrame,
                              df_calificaciones: pd.DataFrame,
                              df_materias: pd.DataFrame) -> Dict:

    df = df_calificaciones.copy()

    df["promedio"] = df[
        ["parcial_1", "parcial_2", "final"]
    ].mean(axis=1)

    return {
        "resumen_general": {
            "total_estudiantes": len(df_estudiantes),
            "promedio_global": df["promedio"].mean(),
            "tasa_aprobacion": (df["promedio"] >= 6).mean() * 100
        },
        "por_semestre": estadisticas_por_semestre(
            df_estudiantes,
            df_calificaciones
        ),
        "fecha_generacion": datetime.now().strftime(
            "%Y-%m-%d %H:%M:%S"
        )
    }
````

````python
def exportar_kardex(boleta: str,
                    kardex: Dict,
                    formato: str = "csv") -> str:

    nombre_archivo = (
        f"kardex_{boleta}_{datetime.now().strftime('%Y%m%d')}.{formato}"
    )

    if formato == "csv":
        kardex["materias"].to_csv(
            nombre_archivo,
            index=False
        )

    elif formato == "json":
        kardex["materias"].to_json(
            nombre_archivo,
            orient="records"
        )

    return nombre_archivo
````
